```python
bucket_name='buc-ket-two'
```

```python
client = boto3.client('s3',
                      aws_access_key_id='AKxxxxxxxxxxxxxxxxxxxx',
                      aws_secret_access_key='qdfCxxxxxxxxxxxxxxxxxxxxxxxxxx',
                      region_name='eu-west-1')
```

```python
response = client.list_objects(
    Bucket = bucket_name
)
```

```python
objectlist = [ content["Key"]for content in response['Contents']]
print(objectlist)
```

```python
header = {
    "data_lesparticuliers":['userid','Titre','Prix','ville','Date']}
```

```python
for objectx in objectlist:
    s3Object= client.get_object(
        Bucket= bucket_name,
        Key = objectx
    )
    for headx in header :
        if headx in objectx :
            hKey = headx
    if "tab" in objectx:
        sep='\t'
    else : sep='|'
    collection=objectx.split("-")[0]
    data = pd.DataFrame(pd.read_csv(s3Object['Body'],delimiter=sep,names=header [hKey],index_col=False))
    records=json.loads(data.T.to_json()).values()
    db[collection].insert_many(records)
    print(collection,'done')
```
