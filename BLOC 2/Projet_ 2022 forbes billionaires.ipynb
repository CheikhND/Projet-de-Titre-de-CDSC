```python
y1 = df["country"].value_counts()
y_label1 =df["country"].value_counts().head(9).keys()
toplabel = [str(x) for x in y_label1]
toplabel.append("Autre")
l = [float(x)for x in y1]
top=[]
s= 0
for i in range (9):
    top.append(l[i])
    s += l[i]
top.append((sum(l) - s))
plt.pie(top, labels=toplabel, startangle=90)
plt.title("Top 10 des pays")
plt.show()
```

# **Partie 2: Pr√©traitement**

**Mise √† l'echelle-Normalisation**

```python
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler
scaler = MaxAbsScaler()
scaler.fit(df[['networth','age']])
scaled = scaler.transform(df[['networth','age']])
scaled_df = pd.DataFrame(scaled)
print(scaled_df)
```

```python
plt.scatter(scaled[:,0], scaled[:,1], color='red')
```

```python
std_scaler = MinMaxScaler().fit(df[['networth', 'age']])
df_std = std_scaler.transform(df[['networth', 'age']])
```

```python
print(df_std)
```

```python
minmax_scale = MinMaxScaler().fit(df[['networth', 'age']])
df_minmax = minmax_scale.transform(df[['networth', 'age']])
```

```python
plt.scatter(df['networth'], df['age'],
            color='green', label='donnees sans transformations', alpha=0.5)
plt.scatter(df_minmax[:,0], df_minmax[:,1],
            color='red', label='min-max scaled [min=0, max=1]', alpha=0.3)
     
plt.scatter(df_std[:,0], df_std[:,1],
            color='blue', label='Standardisation vers la loi Normal X ~ N(0,1)', alpha=0.3)
plt.xlabel('networth')
plt.ylabel('age')
plt.show()
```

```python
x = df[['networth', 'age']]
from sklearn.preprocessing import StandardScaler
mise_a_echelle = StandardScaler()
x = mise_a_echelle.fit_transform(x)
print(x)
```

# **Discr√©tisation en utilisant le groupement (Binning)**

**Cr√©ation d'une nouvelle colonne pour regrouper les fortunes par cat√©gorie: 'petite fortune',  'moyenne',  'grande fortune', 'Tres grande fortune'**

```python
df['networth_bin']=pd.cut(df['networth'],
                        bins = [0,25,50,100,225], 
                        labels = ['petite fortune', 'moyenne', 'grande fortune','Tres grande fortune'])
```

```python
df
```

**Cr√©ation d'une nouvelle colonne pour regrouper les milliadaires par cat√©gorie d'√¢ge:  'jeune',  'junior',  'senior', 'senior+'**

```python
df['age_bin']=pd.cut(df['age'],
                        bins = [0,25,50,70,100], 
                        labels = ['jeune', 'junior', 'senior','senior+'])
```

```python
df
```

```python
y = df['age_bin'].value_counts()
y_label=y.keys()
plt.title("Nombre de milliardaires par cat√©gorie d'√¢ge")
plt.bar(y_label,y)
plt.show
```

```python
y = df['networth_bin'].value_counts()
y_label=y.keys()
plt.title("Nombre de milliardaires par cat√©gorie de fortune")
plt.bar(y_label,y)
plt.show
```
